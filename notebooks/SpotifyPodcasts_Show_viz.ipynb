```python
show_viz_json = defaultdict(dict)
```

```python
import altair as alt



# number of episode
value_counts = filtered_df["show_name"].value_counts()

for value, count in value_counts.items():


    # Create a single-row DataFrame to hold the number
    data = pd.DataFrame({'Number of Episodes': [count]})

    # Create a text chart
    text_chart = alt.Chart(data).mark_text(size=20).encode(
        text='Number of Episodes:Q'  # Q for quantitative
    ).properties(
        title='Number of Episodes'
    )

    show_viz_json[value]["chart1"] = text_chart.to_dict()
```

```python

# filter on show name

# sort by release date

# create line plot of duration on y, and x as episode count

for show_name in filtered_df["show_name"].unique():

    show_df = filtered_df[filtered_df["show_name"] == show_name]

    
```

```python
import altair as alt
import pandas as pd

for show in filtered_df["show_name"].unique():

    df_show = filtered_df[filtered_df["show_name"] == show]

    # Example list of values
    values = filtered_df[filtered_df["show_name"] == show].sort_values(by=["release_date"])["duration"]

    # Convert the list to a DataFrame with an index column, which will serve as the x-axis
    data = pd.DataFrame({
        'episode_number': range(len(values)),
        'duration': values
    })

    # Create the line chart
    line_chart = alt.Chart(data).mark_line().encode(
        x='episode_number:Q',  # Quantitative scale for x-axis
        y='duration:Q'   # Quantitative scale for y-axis
    ).properties(
        title="Duration of each successive episode"
    )

    show_viz_json[show]["chart2"] = line_chart.to_dict()
```

```python
from datetime import datetime

filtered_release_date_df = filtered_df.dropna(subset=['release_date'])
for show in filtered_release_date_df["show_name"].unique():

    df_filtered = filtered_release_date_df[filtered_release_date_df["show_name"] == show]
    min_date = df_filtered["release_date"].min()
    max_date = df_filtered["release_date"].max()

    # Convert strings to datetime objects
    print(show)

    print(min_date)
    print(max_date)

    date1 = datetime.strptime(min_date, '%Y-%m-%d')
    date2 = datetime.strptime(max_date, '%Y-%m-%d')

    # Calculate the difference between dates
    delta = date2 - date1

    # Number of days
    number_of_days = delta.days
    # print(number_of_days)

    avg_frequency = round(number_of_days / df_filtered.shape[0])

    data = pd.DataFrame({'Number of days between episodes': [avg_frequency]})

    # Create a text chart
    text_chart = alt.Chart(data).mark_text(size=20).encode(
        text='Number of Episodes:Q'  # Q for quantitative
    ).properties(
        title='Number of Episodes'
    )

    show_viz_json[value]["chart3"] = text_chart.to_dict()

    
```
