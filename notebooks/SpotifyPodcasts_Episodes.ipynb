```python
import pandas as pd
import altair as alt
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
import json
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('../../podcasts-no-audio-13GB-selected/metadata_with_episode_dates_and_category.tsv',sep='\t')
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d').reset_index(drop=True)
df = df[~df['release_date'].isna()]
df = df[~df['category'].isna()]
```

```python
episode_df = pd.DataFrame(df.iloc[0]).T
episode_df
```

```python
def create_textbox1(current_df, field_name, pretty_field_name,number=False):

    mini_df = pd.DataFrame(current_df[field_name],columns=[field_name])
    if number:
        text_chart = alt.Chart(mini_df).mark_text(baseline="middle").encode(
                text=alt.Text(f"{field_name}:Q", format='.1f'),
        ).properties(
            title=pretty_field_name,  # Add a title
        ).configure_text(
            fontSize=100,  # Adjust the font size
            font='Arial',  # Set the font type
        ).configure_title(
            fontSize=25,  # Adjust the title font size
            font='Arial',  # Set the title font type
        ).configure_view(
            stroke='transparent',  # Set the background color
        )
    else:
        
        #mini_df[field_name] = mini_df[field_name].apply(lambda x: '\n'.join(x[i:i+100] for i in range(0, len(x), 100))).iloc[0]
        
        text_chart = alt.Chart(mini_df).mark_text(baseline="middle").encode(
                text=f"{field_name}:N",
        ).properties(
            title=pretty_field_name,  # Add a title
        ).configure_text(
            fontSize=15,  # Adjust the font size
            font='Arial',  # Set the font type
        ).configure_title(
            fontSize=25,  # Adjust the title font size
            font='Arial',  # Set the title font type
        ).configure_view(
            stroke='transparent',  # Set the background color
        )
    return text_chart





#create_stat1(pub_df).to_dict()
#create_stat1(pub_df)
```

```python
field_name = "episode_description"
pretty_field_name = "Episode Description"
create_textbox1(episode_df, field_name, pretty_field_name)
```

```python
field_name = "show_description"
pretty_field_name = "Show Description"
create_textbox1(episode_df, field_name, pretty_field_name)
```

```python
field_name = "duration"
pretty_field_name = "Episode Duration (min)"
create_textbox1(episode_df, field_name, pretty_field_name,number=True)
```

```python
episodes = {}
field_name1 = "publisher"
pretty_field_name1 = "Publisher"

field_name2 = "show_name"
pretty_field_name2 = "Show Name"

field_name3 = "show_description"
pretty_field_name3 = "Show Description"

field_name4 = "episode_description"
pretty_field_name4 = "Episode Description"

field_name5 = "duration"
pretty_field_name5 = "Episode Duration (min)"

for index, row in df.sample(10000).iterrows():
    episode_df = pd.DataFrame(row).T
    episodes[episode_df.episode_name.iloc[0]] = {
        'stat1': create_textbox1(episode_df, field_name1, pretty_field_name1).to_dict(),
        'stat2': create_textbox1(episode_df, field_name2, pretty_field_name2).to_dict(),
        'stat3': create_textbox1(episode_df, field_name3, pretty_field_name3).to_dict(),
        'stat4': create_textbox1(episode_df, field_name4, pretty_field_name4).to_dict(),
        'stat5': create_textbox1(episode_df, field_name5, pretty_field_name5,number=True).to_dict()
    }

# Specify the file path
file_path = '../visualization_data/episode/episode.json'

# Write the dictionary to a JSON file
with open(file_path, 'w') as json_file:
    json.dump(episodes, json_file)
```
