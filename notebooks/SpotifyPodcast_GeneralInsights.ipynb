```python
# percentage of durations greater than 60, between avg and 59, and less than avg
topten3 = topten.dropna()
topten_duration = topten3
threshold_high = 50
threshold_low = 33
topten_duration['Duration Category'] = pd.cut(topten['duration'], bins=[0, threshold_low, threshold_high, topten['duration'].max()], labels=['Below Threshold', 'Between Thresholds', 'Above Threshold'], right=False)
```

```python
alt.Chart(percentage_df).mark_arc().encode(
    theta="Percentage",
    color="Category"
)
```

# Publisher Info

```python
bypublisher = data.groupby('publisher').size().sort_values(ascending=False)
bypublisher = bypublisher.reset_index()
bypublisher
```

```python
bypublisher.rename(columns={0:"numeps"}, inplace = True)
bypublisher.head(10)
```

```python
#lost of publishers fall off after 2500, then about 700, then less than 500
alt.Chart(bypublisher.sample(5000)).mark_bar().encode(
    x=alt.X('publisher', sort='-y'),
    y= alt.Y('numeps'),

)
```

```python
threshold_lot = 200
threshold_few = 50
bypublisher['Published Category'] = pd.cut(bypublisher['numeps'], bins=[0, threshold_few, threshold_lot, bypublisher['numeps'].max()], labels=['Below Threshold', 'Between Thresholds', 'Above Threshold'], right=False)
```

```python
category_counts = bypublisher['Published Category'].value_counts()
total_episodes = len(bypublisher)
percentage_data = {'Category': category_counts.index, 'Percentage': (category_counts / total_episodes) * 100}
percentage_df = pd.DataFrame(percentage_data)
```

```python
#Showing publish amount trailoff through a pie chart
alt.Chart(percentage_df).mark_arc().encode(
    theta="Percentage",
    color="Category"
)
```
