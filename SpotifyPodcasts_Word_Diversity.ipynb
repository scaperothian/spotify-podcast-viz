# Spotify Podcast Dataset - Calculate Lexical Diversity from Episode Description 

Read in metadata.tsv.  Attempt to look at episode description as a way to measure lexical diversity of the production.

```python
import pandas as pd
```

```python
df = pd.read_csv('podcasts-no-audio-13GB-selected/metadata.tsv', sep='\t')
```

```python
df2 = df.dropna()
def show_word_diversity(row):
    words = row['show_description'].split(' ')
    total_words = len(words)
    #print('total_words')
    return len(set(words)) / total_words

def episode_word_diversity(row):
    words = row['episode_description'].split(' ')
    total_words = len(words)
    return len(set(words)) / total_words

df2['show_word_diversity'] = df2.apply(show_word_diversity, axis=1)
df2['episode_word_diversity'] = df2.apply(episode_word_diversity, axis=1)
```

```python
df2.plot('duration','episode_word_diversity',style='x')
```

# Conclusion

When I look at word diversity with respect to episode description, it is not revealing a lot about episode per se.  With a ratio of 1.0 the description contains all unqiue words.  Whereas with a transcription you would never see a max ratio of 1.0.
