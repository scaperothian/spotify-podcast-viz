# Spotify Podcast Dataset

Read in metadata.tsv.  Attempt to look at the data statistics.  Since most of the data is text, there is not a lot of numerical information out of the box.

```python
import pandas as pd
```

```python
df = pd.read_csv('../../../podcasts-no-audio-13GB-selected/metadata.tsv', sep='\t')
```

```python
# Get the number of NaN values in each column
nan_count = df.isna().sum()

# Get a summary of basic statistics
summary_stats = df.describe()

# Get the number of rows and columns
num_rows, num_columns = df.shape

# Get the names of all columns
column_names = df.columns

# Get the data types of all columns
column_data_types = df.dtypes

# Print the summary
print("Number of NaN values in each column:")
print(nan_count)

print("\nSummary of basic statistics:")
print(summary_stats)

print("\nNumber of rows:", num_rows)
print("Number of columns:", num_columns)

print("\nColumn names:")
print(column_names)

print("\nData types of columns:")
print(column_data_types)
```

```python
print(f"Unique Values By Column.  Number of rows: {df.shape[0]}")
for col_name in df.columns:
    print(f"{col_name}: {df[col_name].unique().shape[0]}")
```
